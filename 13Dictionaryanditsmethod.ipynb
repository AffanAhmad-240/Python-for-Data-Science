# Dictionary
- Dictionary is an inordered, mutable collection of key value pairs. It allows fast reterival of data.

```python
dic = {}
```

```python
type(dic)
```

```python
teacher = {
    "Name" : "Hafsa",
    "Subject" : "SDA",
    "Salary" : 120000
}
```

```python
print(teacher)
```

```python
teacher["Salary"]
```

```python
teacher["Salary"] = 130000     #updates salary in original dictionary
```

```python
print(teacher)
```

```python
teacher["Age"] = 28
```

```python
print(teacher)
```

# Dictionary Methods

```python
teacher.keys()
```

```python
teacher.values()
```

```python
teacher.items()
```

```python
teacher.get("Name")
```

```python
r = teacher.get("Namerz")     # get() method return none if key is not present in dictionary.
```

```python
print(r)
```

```python
# we can also give default argument which is shown when key is not present in dictionary. e.g:
teacher.get("Namerf" , "Key not found")
```

```python
teacher.pop("Age")
```

```python
print(teacher)
```

```python
teacher.popitem()        #remove and return and last inserted key-value pair.
```

```python
print(teacher)
```

```python
teacher.setdefault("Awq" , "Unknown")          # check wether this key is present in dict or not. If not insert this new key
                                               # and sets its value as default (e.g: Unknown in this case)
```

```python
dict2 = teacher.copy()
```

```python
dict2
```

```python
dict2.clear()
```
