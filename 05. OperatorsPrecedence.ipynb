# One golden rule

- When two or more exponentiation (power, ) came in equation, it is solved from right to left.
- 2 power 3 power 2 = 2**9 = 512
- When (* and /) or (+ and -) came in same equation, it is solved from left to right.
- Bcz * and / has same precedence which come first, solved first. Same case for + and -
