```python
s1 = set()     # empty set
```

```python
type(s1)
```

```python
s2 = {2, 3, 6, 8, 9, 2, 3, 9}
```

```python
s2  # it only shows unique values
```

# Set operations

```python
s2.union(s3)              # return new set, original sets remain unchanged
```

```python
s2.intersection(s3)
```

```python
U = {1,2,3,4,5,6}
A = {4,5}
U-A
```

```python
A.issubset(U)
```

```python
U.issuperset(A)
```

```python
U.difference(A)
```

```python
V = {1,2,3,4,5,6}
B = {4,5,7,8}
V.symmetric_difference(B)
```
