- In Python, format() and F-strings are used for string formatting.
- It means putting variables and values inside text in a clean, readable way.
- format() method is an old used to format string.
- e.g: print("My name is {}. I am {} years old".format(name, age))
- F-strings are the modern way used for string formatting.
- print(f"My name is {name}. I am {age} years old.")

```python
name = "AA"
age = 21
print("My name is {}. I am {} years old.".format(name, age))
```

```python
name = "AA"
age = 21
print(f"My name is {name}. I am {age} years old.")
```

```python
num = 181.08621
f"{num:.2f}"
```

```python
rs = 1865942921
f"{rs:,}"
```

# Padding and alignment

```python
print(f"{'Python':<10}")    # left aligned
print(f"{'Python':^10}")    # Center aligned
print(f"{'Python':>10}")    # Right aligned
```
