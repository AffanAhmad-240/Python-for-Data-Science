# Calculator

```python
a = int(input("a: "))
```

```python
b = int(input("b: "))
```

```python
sign = input()
```

```python
match sign:
    case "+":
        print("sum: ", a+b)
    case "-":
        print(f"subtract: {a-b}")
    case "*":
        print("Multiply: ", a*b)
    case "/":
        print(f"Divide: {a/b}")
    case "**":
        print("Exponentiation: ", a**b)
    case _:
        print("Not a sign")
```
