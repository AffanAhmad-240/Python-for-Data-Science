# Using pass, break and continue statements in for, while loop and check how it works

```python
items = ['1. Bananas' , '2. Apples', '3. Mangoes' , '4. Orange' , '5. Tomatoes', '6. Chilly']
for item in items:
    if item == "5. Tomatoes":
        break    # loop terminates immediately
    print(item)
```

```python
# Print even numbers using continue statement
i = 0
while i <= 10:
    if i % 2 != 0:
        i = i+1
        continue    # skip this iteration and move to 
    print(i)
    i = i+1
```

```python
# without pass statement
for i in range(4):

print("Program End")
```

```python
# with pass statement
for i in range(4):
    pass             # Don't do anything, just move to the next line
print("Program End")
```
