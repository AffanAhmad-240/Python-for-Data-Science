# if else statements

```python
a = 8
if (a > 50):
    print("Greater than 50")
else:
    print("Lesser")
```

```python
x = 26
if (x > 50):
    print("Greater than 50")
elif (x>25):
    print("Greater than 25")
elif (x>12.5):
    print("Greater than 12.5")
else:
    print("Lesser than or equal to 12.5")
```

# Functions

```python
def power(a,b):
    return a**b
```

```python
power(2,3)
```

```python
def power(a,b=2):   # with default parameter
    return a**b
```

```python
power(4)
```
